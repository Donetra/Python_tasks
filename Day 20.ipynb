```python
c=(4,5,6,7)
d=(6,7,8,9)
e=(1,2,3,4)
f=(0,4,8,1)
```

```python
plt.plot(c,d,label='firstline')
plt.plot(e,f,label='secondline')
plt.xlabel("abc")
plt.ylabel("xyz")
plt.title("aaa")
plt.legend()
plt.show
```

```python
plt.plot(c,d,label='firstline',color = 'green', linestyle = '--', linewidth = 0.5)
plt.plot(e,f,label='secondline',color = 'red', linestyle = '-', linewidth = 1)
plt.xlabel("abc")
plt.ylabel("xyz")
plt.title("aaa")
plt.legend()
plt.show()
```

```python
plt.plot(c,d,label='firstline',color = 'green',linestyle = 'solid', marker = 'o',markerfacecolor = 'red', markersize = 12)
plt.plot(e,f,label='secondline',color = 'red', linestyle = 'solid', color = 'green',marker = 'o',
         markerfacecolor = 'red', markersize = 12linewidth = 1)
plt.xlabel("abc")
plt.ylabel("xyz")
plt.title("aaa")
plt.legend()
plt.show()
```
